# Find HHI 

We get the portion data per national level. In order to get numbers of GDP, we need some simple calculation. But, overall that way more efficient than scraping all province data one by one.

```python
lq.head()
```

```python
indo =  pd.read_excel("dataugm.xlsx", sheet_name="Indonesia").iloc[:,:18]
indo.head()
```

```python
h = {}
for i in lq.tahun.unique():
    x = lq[lq["tahun"] == i].iloc[:, 1:-2]
    y = indo[indo["tahun"] == i].iloc[:,1:].values
    o = x.mul(y, axis="columns")
    o["tahun"] = i
    o["provinsi"] = lq.provinsi
    h[i] = o
```

```python
satu = h[2010]
for x in lq.tahun.unique()[1:]:
    satu = pd.concat([satu, h[x]], axis=0)
```

```python
satu.head()
```

```python
tahun = satu["tahun"]
provinsi = satu["provinsi"]
del satu["tahun"] , satu["provinsi"]
jika = satu.iloc[:,:]
jika["jumlah"] = jika.sum(axis=1)
for i in jika:
    jika[i] = (jika[i] / jika.jumlah) * 100
```

```python
jika =  jika ** 2
```

```python
jika.head(15)
```

```python
satu["hhi"] = jika.iloc[:,:-1].sum(axis=1)
satu["tahun"] = tahun
satu["provinsi"] = provinsi
```

```python
satu = satu.set_index(["provinsi", "tahun"])
satu.head(10)
```

# Comparative Advantage and Concentration

```python
jadi_lq["hhi"] = satu["hhi"]
jadi_lq["primary_portion"] = satu.iloc[:,0] + satu.iloc[:,1]
jadi_lq.head()
```

```python
%matplotlib inline
jadi_lq.hhi.plot(kind="box")
```

```python
semifinal = jadi_lq[jadi_lq["hhi"] <= 1500]
del semifinal["hhi"] , 
print (jadi_lq.shape)
print (semifinal.shape)
```

```python
final = semifinal[semifinal > 1.1]
```

```python
finall = final.reset_index().groupby("provinsi").mean()
more = finall["primary_portion"]
del finall["tahun"] , finall["primary_portion"]
finall["num_of_potential_sector"] = finall.notnull().sum(axis=1)
```

```python
finall.shape
```

```python
finall = finall.sort_values("num_of_potential_sector", ascending=False)
finall.num_of_potential_sector.plot(kind="barh")
```
